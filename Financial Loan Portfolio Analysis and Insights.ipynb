```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\Owner\Downloads\financial_loan.csv")
df
```

```python
# QUESTION 1: Shape, columns, and data types 
print("Shape:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nFirst few rows:\n", df.head())
```

```python
# ===== QUESTION 2: Missing values =====
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({
    'Missing_Count': missing_values,
    'Missing_Percentage': missing_percentage
})
print("\nMissing values:\n", missing_df)
```

```python
# ===== QUESTION 3: Summary statistics =====
print("\nSummary statistics:\n", df.describe())
```

```python
# ===== QUESTION 4: Distribution of loan status =====
# Replace 'loan_status' with your actual column name
loan_status_counts = df['loan_status'].value_counts()
loan_status_percentage = df['loan_status'].value_counts(normalize=True) * 100
status_df = pd.DataFrame({
    'Count': loan_status_counts,
    'Percentage': loan_status_percentage
})
print("\nLoan status distribution:\n", status_df)
```

```python
# ===== QUESTION 5: Loans > $10,000 and average interest rate =====
# Replace column names with your actual column names
loans_above_10k = df[df['loan_amount'] > 10000]
print("\nNumber of loans > $10,000:", len(loans_above_10k))
print("Average int_rate:", loans_above_10k['int_rate'].mean())
```

```python
# ===== QUESTION 6: Top 10% income earners =====
income_threshold = df['annual_income'].quantile(0.90)
top_10_percent = df[df['annual_income'] >= income_threshold]
print("\nTop 10% income earners count:", len(top_10_percent))
```

```python
# QUESTION 7: Debt-to-Income ratio and correlation =====
df['debt_to_income_ratio'] = df['loan_amount'] / df['annual_income']
print("\nDebt-to-Income ratio statistics:\n", df['debt_to_income_ratio'].describe())
```

```python
# ===== QUESTION 8: Outliers using IQR method =====
Q1 = df['loan_amount'].quantile(0.25)
Q3 = df['loan_amount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['loan_amount'] < lower_bound) | (df['loan_amount'] > upper_bound)]
print("\nNumber of outliers:", len(outliers))
print("Percentage of outliers:", (len(outliers) / len(df)) * 100, "%")
print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)
```

```python
# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python

# ===== GRAPH 1: Distribution of Loan Amounts (Histogram) =====
plt.figure(figsize=(10, 6))
plt.hist(df['loan_amount'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Loan Amounts', fontsize=16, fontweight='bold')
plt.xlabel('Loan Amount ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
# ===== GRAPH 2: Loan Status Distribution (Bar Chart or Pie Chart) =====
# Option A: Bar Chart
plt.figure(figsize=(10, 6))
loan_status_counts = df['loan_status'].value_counts()
loan_status_counts.plot(kind='bar', color=['#2ecc71', '#e74c3c', '#f39c12'], edgecolor='black')
plt.title('Loan Status Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Loan Status', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
# Option B: Pie Chart
plt.figure(figsize=(8, 8))
loan_status_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, 
                        colors=['#2ecc71', '#e74c3c', '#f39c12'])
plt.title('Loan Status Distribution', fontsize=16, fontweight='bold')
plt.ylabel('')  # Remove default ylabel
plt.tight_layout()
plt.show()
```

```python
# ===== GRAPH 3: Relationship between Income and Loan Amount (Scatter Plot) =====
plt.figure(figsize=(10, 6))
plt.scatter(df['annual_income'], df['loan_amount'], alpha=0.5, c='coral', edgecolors='black', s=50)
plt.title('Relationship between Income and Loan Amount', fontsize=16, fontweight='bold')
plt.xlabel('Income ($)', fontsize=12)
plt.ylabel('Loan Amount ($)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# GRAPH 4: Box Plot - Loan Amount by Loan Status
plt.figure(figsize=(10, 6))
df.boxplot(column='loan_amount', by='loan_status', grid=False)
plt.title('Loan Amount Distribution by Status', fontsize=16, fontweight='bold')
plt.suptitle('')  # Remove automatic title
plt.xlabel('Loan Status', fontsize=12)
plt.ylabel('Loan Amount ($)', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# GRAPH 5: Correlation Heatmap
plt.figure(figsize=(10, 8))
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', square=True, linewidths=1)
plt.title('Correlation Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
